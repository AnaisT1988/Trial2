```python
import pandas as pd
import requests
import credentials
```

```python
# server
base_URL="http://10.73.58.47/"
```

```python
# Authorization 

URL = base_URL +"cst-api/v1/infrastructure/authorization/login"

headers = {
"accept": "application/json",
"Content-Type": "application/json"
}

Credentials={
  "Credentials": {
    "Username": credentials.username,
    "Password": credentials.password
  }
}


resp = requests.post(URL,headers=headers,json=Credentials)
#tk = json.loads(resp.text)
j=resp.json()
token=j['authorizationContext']['accessToken']

if resp.status_code != 200:
    print('error: ' + str(resp.status_code))
else:
    #print('token: ' + str(token))
    print('Success')
```
