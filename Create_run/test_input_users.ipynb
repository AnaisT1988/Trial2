```python

for i in df['CAS Number']:
    if i!='-':
        data= [{"AttributeDefinition": {
        "CstId": "345FAE2E-54E4-4D62-A693-8A8C00A031F8"},
        "Values": [{"String":i}],
        }
        ]

        resp_2 = requests.post(URL_2,headers=headersAPI,json=data,verify=True)

        if resp_2.status_code == 200:
            tk_2=resp_2.json()
            ex=tk_2['result'][0]
            ex_2=ex['attributes']

            index=df[df['CAS Number'].isin([i])].index

            for j in range (0,len(ex_2)-1):
                if ex_2[j]['attributeDefinition']['id'] == 95:
                    df.loc[index,'Melting point']= ex_2[j]['flattenedValue']
                elif ex_2[j]['attributeDefinition']['id'] == 169:
                    df.loc[index,'Molar mass']= ex_2[j]['flattenedValue']
                
                #elif ex_2[i]['attributeDefinition']['id'] == 33:
                #  df['CAS NUMBER']= ex_2[i]['flattenedValue']
                    
                elif ex_2[j]['attributeDefinition']['id'] == 139:
                    df.loc[index,'Dispense Tool Volumetric']= ex_2[j]['flattenedValue']

                elif ex_2[j]['attributeDefinition']['id'] == 138:
                    df.loc[index,'Dispense Tool Gravimetric']= ex_2[j]['flattenedValue']
                    
                elif ex_2[j]['attributeDefinition']['id'] == 254:
                    df.loc[index,'Internal Standard']= ex_2[j]['flattenedValue']
                    
                elif ex_2[j]['attributeDefinition']['id'] == 198:
                    df.loc[index,'Melting duration']= ex_2[j]['flattenedValue']
                
                elif ex_2[j]['attributeDefinition']['id'] == 28:
                    df.loc[index,'Dispensable']= ex_2[j]['flattenedValue']
                    
                elif ex_2[j]['attributeDefinition']['id'] == 59:
                    df.loc[index,'Density']= ex_2[j]['flattenedValue']
                    
                else:
                    continue
        else:
            pass
    else:
        pass
```

```python
def get_att(base_URL,token,data_T):

    URL_2= base_URL +"/cst-api/v1/product-management/articles/get-articles-by-attributes-values"

    headersAPI = {
    "accept": "application/json",
    'Authorization': 'Bearer '+ token,
    }
    
    for i in data_T['CAS Number']:
        if i!='-':
            data= [{"AttributeDefinition": {
            "CstId": "345FAE2E-54E4-4D62-A693-8A8C00A031F8"},
            "Values": [{"String":i}],
            }
            ]

            resp_2 = requests.post(URL_2,headers=headersAPI,json=data,verify=True)

            if resp_2.status_code == 200:
                tk_2=resp_2.json()
                ex=tk_2['result'][0]
                ex_2=ex['attributes']

                index=data_T[data_T['CAS Number'].isin([i])].index

                for j in range (0,len(ex_2)-1):
                    if ex_2[j]['attributeDefinition']['id'] == 95:
                        data_T.loc[index,'Melting point']= ex_2[j]['flattenedValue']
                    elif ex_2[j]['attributeDefinition']['id'] == 169:
                        data_T.loc[index,'Molar mass']= ex_2[j]['flattenedValue']
                    
                    #elif ex_2[i]['attributeDefinition']['id'] == 33:
                    #  data_T['CAS NUMBER']= ex_2[i]['flattenedValue']
                        
                    elif ex_2[j]['attributeDefinition']['id'] == 139:
                        data_T.loc[index,'Dispense Tool Volumetric']= ex_2[j]['flattenedValue']

                    elif ex_2[j]['attributeDefinition']['id'] == 138:
                        data_T.loc[index,'Dispense Tool Gravimetric']= ex_2[j]['flattenedValue']
                        
                    elif ex_2[j]['attributeDefinition']['id'] == 254:
                        data_T.loc[index,'Internal Standard']= ex_2[j]['flattenedValue']
                        
                    
                    elif ex_2[j]['attributeDefinition']['id'] == 28:
                        data_T.loc[index,'Dispensable']= ex_2[j]['flattenedValue']
                        
                    elif ex_2[j]['attributeDefinition']['id'] == 59:
                        data_T.loc[index,'Density']= ex_2[j]['flattenedValue']
                        
                    else:
                        continue
            else:
                pass
        else:
            pass

    return data_T   
```
